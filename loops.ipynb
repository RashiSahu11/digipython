```python
fruits=['apple','banana','cherry','kiwi','onion','mango','orange']
for x in fruits:
    if x=='onion':  
       break
    print(x)
```

```python
fruits=['apple','banana','cherry','onion', 'kiwi','mango']
for x in fruits:
    if x=='onion':
        continue
    print(x)
```
