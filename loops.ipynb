```python
#reverse of a number
num=int(input('Enter a number:'))
orig=num
reverse=0

while num>0:
    digit=num%10
    reverse=reverse*10+digit
    num=num//10

print('Original number',orig)
print('Reversed number',reverse)    
```

```python
num=int(input("Enter the number:")) #pallandrome number
orig=num
reverse=0

while num>0:
    digit=num%10
    reverse=reverse*10+digit
    num=num//10
if orig==reverse:
    print("Number is palandrom no.") 
else:
    print('no. is not planadrom')       
```

```python
#Armstrong number
num=int(input("Enter a number"))
orig=num
digits=len(str(num))

sumofpowers=0

while num>0:
    digit=num%10
    sumofpowers +=digit ** digits
    num=num//10

if orig == sumofpowers:
    print("Armtrong number")
else:
    print("not armstorng number")    
```
